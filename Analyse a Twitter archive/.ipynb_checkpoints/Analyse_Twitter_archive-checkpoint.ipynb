## 1. Introduction

The dataset which will be wrangling (and analyzing and visualizing) is the tweet archive of Twitter user @dog_rates, also known as **WeRateDogs**. **WeRateDogs** is a Twitter account that rates people's dogs with a humorous comment about the dog. These ratings almost always have a denominator of 10. The numerators, though? Almost always greater than 10. 11/10, 12/10, 13/10, etc. Why? Because "they're good dogs Brent." WeRateDogs has over 4 million followers and has received international media coverage.

**The goal**: wrangle WeRateDogs Twitter data to create interesting and trustworthy analyses and visualizations. The Twitter archive is great, but it only contains very basic tweet information. Additional gathering, then assessing and cleaning is required for "Wow!"-worthy analyses and visualizations.

## 4. Cleaning Data


```python
# Make copies of original pieces of data
twitter_clean = twitter.copy()
image_pred_clean = image_pred.copy()
tweet_clean = tweet.copy()
```

### Quality Issues

### Issue #1: Change some weird names appears (a, an, the).

#### Define

List out weird names and then change those weird names into 'None' 

(do not delete because sometimes the owner of dog has yet to decide the name for his dog)

#### Code

```python
#List out names
twitter_clean.groupby('name').size()
```

```python
#It seems that weird names do not start with capital letter in this scenario
# Check names starts with normal (not capital) letter.
weird_name = twitter_clean['name'][twitter_clean['name'].str[:1].str.isupper() ==False].unique()
weird_name
```

```python
# Change these weird names into 'None'
twitter_clean['name'].replace(weird_name, 'None', inplace=True)
```

#### Test

```python
check = []

for x in range(0,len(weird_name)):
    check.append(sum(twitter_clean['name'] == weird_name[x]))

check
```

### Issue #2: A tweet (observation) is a not orginal (retweets, replies)

#### Define
If a tweet (observation) has value in either one of these (retweeted_status_id; retweeted_status_user_id, retweeted_status_timestamp; in_reply_to_status_id in_reply_to_user_id columns), that tweet is a not orginal and thus will be removed.

#### Code

```python
column = ['retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp', 'in_reply_to_status_id', 'in_reply_to_user_id']
for x in column:
    twitter_clean = twitter_clean[twitter_clean[x].isnull()]
```

#### Test

```python
twitter_clean.info()
```

```python
# After cleaning retweets and replies, we can delete these rows so that the table looks neater
twitter_clean.drop(column, axis = 1, inplace = True)
```

```python
twitter_clean.info()
```

### Issue #7: Inconsistent format of value in the column p1, p2, p3.

```python
# Capitalize first letter
image_pred_clean[['p1', 'p2', 'p3']] = image_pred_clean[['p1', 'p2', 'p3']].apply(lambda x: x.str.capitalize())

# Replace '-' by white space
image_pred_clean[['p1', 'p2', 'p3']] = image_pred_clean[['p1', 'p2', 'p3']].apply(lambda x: x.str.replace('_', ' '))
```

### Tidiness Issue

### Issue #1: there are 4 seperate dog 'stage' columns

#### Define

To make the table cleaner, I will combine four columns (doggo', 'floofer', 'pupper', 'puppo') together under a common column 'stage'.

#### Code

```python
# Delete 'None' string in all 4 dog stages for easier combination.
twitter_clean['doggo'] = twitter_clean['doggo'].replace('None','')
twitter_clean['floofer'] = twitter_clean['floofer'].replace('None','')
twitter_clean['pupper'] = twitter_clean['pupper'].replace('None','')
twitter_clean['puppo'] = twitter_clean['puppo'].replace('None','')
```

```python
# Create new column 'stage' containing values from 4 dog stage columns
twitter_clean['stage'] = twitter_clean['doggo']+ twitter_clean['floofer']+twitter_clean['pupper']+twitter_clean['puppo']
```

```python
# Replace '' (white space) value in the stage column by 'None'
twitter_clean['stage'] = twitter_clean['stage'].replace('', 'None')
```

```python
# Check values in the new column 'stage'
twitter_clean['stage'].value_counts()
```

It looks like that one dog can be assigned into two stages (e.g: doggo & floofer)

```python
# Add a comma between two stage names if applicable
twitter_clean[twitter_clean['stage']=='doggofloofer'] = twitter_clean[twitter_clean['stage']=='doggofloofer'].replace('doggofloofer', 'doggo, floofer')
twitter_clean[twitter_clean['stage']=='doggopupper'] = twitter_clean[twitter_clean['stage']=='doggopupper'].replace('doggopupper', 'doggo, pupper')
twitter_clean[twitter_clean['stage']=='doggopuppo'] = twitter_clean[twitter_clean['stage']=='doggopuppo'].replace('doggopuppo', 'doggo, puppo')
```

#### Test

```python
twitter_clean['stage'].value_counts()
```

```python
# Delete 4 old stage columns

twitter_clean = twitter_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1)
```

### Issue #2: 'text' column contains 2 pieces of information

#### Define

Because the attached link in the 'text' column is already presented in 'expanded_urls'; hence, I will delete it.

#### Code

```python
twitter_clean['text'] = twitter_clean['text'].str.replace(r'https?://[^\s<>"]+|www\.[^\s<>"]+', ' ')
```

```python
twitter_clean['text'].sample(4)
```

## 5. Storing Data
Save gathered, assessed, and cleaned master dataset to a CSV file named "twitter_archive_master.csv".

```python
# Merge three tables together
twitter_archive_master = pd.merge(twitter_clean, image_pred_clean,
                       how='left', left_on='tweet_id', right_on='tweet_id')

twitter_archive_master = pd.merge(twitter_archive_master, tweet_clean,
                       how='left', left_on='tweet_id', right_on='id')
```

```python
twitter_archive_master.info()
```

```python
# Drop the 'id' column in the final table
twitter_archive_master = twitter_archive_master.drop('id', axis=1)
```

Generally, twitter_archive_master table apparently contains many missing values, especially from columns belong to image_pred table.
This result may come from the fact that predictive algorithms of WeRateDog did not perform properly or pictures are not posted by WeRateDog?

Due to unlimited access to the predictive algorithm, I will check the former hypothesis first 

```python
# Check the urls of tweets that are not recognized by predictive algorithm
twitter_archive_master['expanded_urls'][twitter_archive_master['p1'].isnull()].sample(15)
```

```python
# Check the urls of tweets that are recognized by predictive algorithm
twitter_archive_master['expanded_urls'][twitter_archive_master['p1'].isnull() == False].sample(15)
```

My stated hypothesis is seemingly correct when the predictive algorithm only performs on pictures posted by WeRateDog (expanded_urls contain 'https://twitter.com/dog_rates')

```python
# Check if there are any pictures posted by WeRateDog were missed by the predictive algorithm.
check1 = twitter_archive_master[twitter_archive_master['p1'].isnull()]
check1[['tweet_id','expanded_urls']][check1['expanded_urls'].str.contains('https://twitter.com/dog_rates',case=True)]
```

The algorithm missed 3 posts by WeRateDog. However, this is reasonable because these 3 posts, after being checked, does not present any pictures of dog. So it is safe to delete all rows with missing values in p1 (or any columns belong to the old image_pred table).

```python
# Delete observations with missing values in the p1 column

twitter_archive_master = twitter_archive_master[twitter_archive_master['p1'].isnull()==False]
```

```python
twitter_archive_master.info()
```

```python
# Save the final table in csv file
twitter_archive_master.to_csv('twitter_archive_master.csv', index = False)
```

#### 1. What are top 10 common names of dogs reported by WeRateDog?


```python
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

# Create a dataset for visualization
data = pd.DataFrame(twitter_archive_master['name'][twitter_archive_master['name'] != 'None'].value_counts()[:10])
data = data.reset_index()
data.columns = ['Name', 'Count']


# Visualize result
fig = plt.figure(figsize=(12, 8))
sb.barplot(x="Name", y="Count", data=data)
plt.title('Top 10 common dog name')
```

```python
twitter_archive_master['name'].nunique()
```

#### 2. The rank of dog breed in terms of average favorite counts?


```python
# Extract observations corresponding to each breed
doggo = twitter_archive_master[twitter_archive_master['stage'].str.contains('doggo')]
puppo = twitter_archive_master[twitter_archive_master['stage'].str.contains('puppo')]
floofer = twitter_archive_master[twitter_archive_master['stage'].str.contains('floofer')]
pupper = twitter_archive_master[twitter_archive_master['stage'].str.contains('pupper')]
```

```python
# Create a dataframe to store values
name =['doggo', 'puppo', 'floofer', 'pupper']
avg_fav = [doggo['favorite_count'].mean(), puppo['favorite_count'].mean(), floofer['favorite_count'].mean(), pupper['favorite_count'].mean()]
data = pd.DataFrame(list(zip(name, avg_fav)))
data.columns = ['Name', 'Avg_Fav']
data_sort = data.sort_values('Avg_Fav', ascending=False)
```

```python
# Plot the resuslt
fig = plt.figure(figsize=(12, 8))
sb.barplot(x="Name", y="Avg_Fav", data=data_sort)
plt.title('Rank of dog breed by average favorite count')
```

#### 3. Relationship between favorite counts and retweets number?

```python
fig = plt.figure(figsize=(12, 8))
sb.regplot(data = twitter_archive_master, x = 'favorite_count', y = 'retweet_count')
plt.xlabel('Favorite Count')
plt.ylabel('Retweet Count')
plt.title('Relationship between favvorite and retweets count')
```

```python
np.corrcoef(twitter_archive_master['favorite_count'], twitter_archive_master['retweet_count'])
```

## 7. Conclusions:
1. We have top 10 common names for dog: Charlie, Oliver, Cooper, Lucy, Tucker, Penny, Sadie, Winston, Lola, and Daisy. In total, there are 913 different names of dog recorded in the data frame. The most famous name, Charlie, is founded in only 11 tweets. However, 'None' value accounts for a third of total observations, 622 out of 1971. These findings suggest that not every post from WeRateDog includes dog‚Äôs name or that the algorithm of extracting dog‚Äôs name has yet to perform properly.


2. People tend to love the ‚Äòpuppo‚Äô dogs when tweets related to this breed received more than 22 000 favorite counts. This number is approximately three times that of ‚Äòpupper‚Äô dog class ‚Äì the least favorite one. One of the main reasons justifying this finding is that the tweet with highest favourite counts, 132810, calls the name of ‚Äòpoppo‚Äô class. This result is quite surprising since the number of ‚Äòpupper‚Äô-related tweets are the highest amongst 4 breeds. On the other hand, there are only 23 tweets mentioning puppo, making this class rank penultimate in the list.



3. There is a pretty strong positive relationship between retweet count and favorite count. The Pearson correlation coefficient between these two variables is up to 0,92. The closer the coefficient score is to 1, the stronger the correlation between two feature is. This finding is quite obvious because a post, if are retweeted many times, will catch attention of more users and thus their favorite, and vice versa. In the reverse way, a post, if receives many favorites from people, will potentially go viral and be retweeted multiple times.
